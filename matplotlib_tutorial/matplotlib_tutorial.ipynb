# Anatomy of a plot

#### 3 Important "Classes"

- matplotlib.pyplot
    - Instance based class without callbacks
    - Attributes match most of matlab attributes


- Axes
    - Instance based and supports callbacks
    - Modify "sample" attribute with "set_sample"
    - Contains the does not modify the plotted instance but the axis, ticks, titles, etc.
   

- Figure container
    - Containter for all plotting elements   



The differences in using plyplot (**plt.some_attribute()**) and axes (**axes.set_some_attribute()**) do change the way you plot and think about plotting.
For anything that only displays one figure, using **plt** is easiest.
For anything that uses subplots, calling the **axes** class is more robust and easier (although more wordy).
As for the **Figure** container, that is used in special circumstances that I will demonstrate at the end of the talk.

# Instance based plotting 

For example here we will plot subplots using the axes method. 
We first define some x-data then plug it into the sine and cosine functions from the numpy package.
For formality, we call the figure function from pyplot because if we need to we can adjust the size, background color, etc.
If you're a matlab user you may recognize the the convention of calling a subplot using the rows, columns, then placement.
After plotting these two functions we add some labels. 
Here you notice that the same function *plt.ylabel* is called twice yet does two different things.
There are no brackets to separate the two subplots so instead, plt.any_function refers to the last instanced subplot.
In this case, we must contain all functions we want for each subplot before the next one is called.
Admittedly this is easier to read what happens to each graph, but for more complex/clever graphs, our next method is much easier to handle.


```python
import matplotlib.pyplot as plt
import numpy as np

# ---- Simple Trig Functions ----
x = np.linspace(0,2*np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)

# ---- Create Figure ----
plt.figure()
 
# subplot(a,b,c) or subplot(abc)
# a = rows, b = columns, c=this subplot's place on the figure
plt.subplot(211)
plt.plot(x,y1)
plt.ylabel('sin(x)')

plt.subplot(212)
plt.plot(x,y2) 
plt.ylabel('cos(x)')
plt.xlabel('x-axis')
plt.show()

plt.tight_layout()
```

# Using the Axes Class

Below we plot the exact same functions, but this time we prerender the subplots by defining the rows and columns.
The outputs for this are exactly the same as the previous example. 
Only now, we were free to group the y-labels together.

```python
import matplotlib.figure as fg
import numpy as np

# ---- Simple Trig Functions ----
x = np.linspace(0,2*np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)

# ---- Create Figure ----
fig, axes = plt.subplots(nrows=2,ncols=1)
 
# plot figures
axes[0].plot(x,y1) 
axes[1].plot(x,y2)

# set y-labels
axes[0].set_ylabel('sin(x)')
axes[1].set_ylabel('cos(x)')

```

# Modify the figure

- Figure Resizing and Colors
- Line width, styles, and colors
- LaTeX

Like I said earlier, when we instantiate our figures before plotting, we can easily change the properties of the figure.
In this block of code I pretyped some properties for us to play with: figure size, dpi, and face/edge color.


On top of that we can edit the rendering of the lines by defining the line width, line style, and color.
In the comments are some common (if not all) linestyles available.


Additionally you can edit the colors which I have done so on both graphs.
The comments give you a list of valid colors plus a neat feature which allows you to pull from a group of colors.
The second graph uses the color red, but not rgb red.
The color was taken from a survey put out by the xkcd comic where they essentially asked which color looked *most* red.

Lastly, python has a simple LaTeX interpreter which is nice for simple equations and expressions. 
Note that you must use the argument **r'string'** to indicate the string is *raw*.

```python
import matplotlib.figure as fg
import numpy as np

# ---- Simple Trig Functions ----
x = np.linspace(0,2*np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)

# ---- Create Figure ----
fig, axes = plt.subplots(nrows=2,ncols=1,
                         figsize=(2,2),
                         dpi=200,
                         facecolor='w',
                         edgecolor='w',
                        )
 
# plot figure
# Linestyles= '-'(default), '--', ':', '-.' 
axes[0].plot(x,y1,linewidth = 5,linestyle='-.',color='k') 

# Colors
#     r,g,b,w
#     c,m,y,k
#     greysacle: 0.0 - 1.0 (white - black)
#     Hex values accepted
#     'color_group_example:color_example'
#         'tab:color','xkcd:color'
axes[1].plot(x,y2,linestyle='--',color='xkcd:red')

# set y-labels
axes[0].set_ylabel('sin(x)')
axes[1].set_ylabel('cos(x)')

# set x-label (Simple Latex supported)
axes[1].set_xlabel(r'$\theta +  \alpha \omega t$')

# erase x-ticks in first subplot because they're redundant
axes[0].set_xticklabels([])
```

# Markers and Scatter Plots 

Only use plt.scatter for 3D-Data! Why? plt.plot is faster!

Markers and scatter plots are often necessary, but there is however one caveat to when you should be using the function **plt.scatter()**
The function **plt.scatter()** not only plots a point everywhere you define, it also attempts to define a size for each marker based on a third argument. This can make plots render extremely slow for large datasets.

Instead use **plt.plot(x,y,linestyle='',marker='.')**. 
This erases the lines and sets a marker instead. 
From a design perspective, this is confusing.
But like I said, it shouldn't be a huge problem until you encounter large datasets.


A list of *some* markers is in the comments.

```python
import matplotlib.figure as fg
import numpy as np

# ---- Simple Trig Functions ----
x = np.linspace(0,2*np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)

# ---- Create Figure ----
fig, axes = plt.subplots(nrows=2,ncols=1,
                         figsize=(5,5),
                         dpi=200,
                         facecolor='w',
                         edgecolor='w',
                        )
 
# plot figure
# Linestyles= '-'(default), '--', ':', '-.' 
axes[0].plot(x,y1,linestyle='-.') # change linewidth

# ---- Markers: ----
#     .,+,x,o,D,d,8,s,[1-7]
axes[1].plot(x[::4],y2[::4],linestyle='',marker='s')

# set y-labels
axes[0].set_ylabel('sin(x)')
axes[1].set_ylabel('cos(x)') # only plot every 4th point to show markers

# set x-label (Simple Latex supported)
axes[1].set_xlabel(r'$\theta + \omega t$')

# erase x-ticks in first subplot because they're redundant
axes[0].set_xticklabels([])
```

# Contour Maps
###### Credit matplotlib.org


Contour maps are one of the most pleasing looking plots to me and matplotlib has some nice ones. 
To plot an unfilled contour map you simply use the function **contour()**.
If you would like a filled contour use **contourf()**.

Here we create some data and arrange it properly using a meshgrid of the needed data.
Then we define the subplots before the graphs as usual.
Now we plot the **unfilled contour**
And here we actually assign the plotted instance to a variable in order to set the cool inline labels on the unfilled contour graph.

Then we set the **filled contour** and also assign it to a variable.
Now this is where we use the **figure** instance and essentially all I use it for.
This sets up the color bar and requires a call to both what *subplot* instance you work on, and the *plotted* instance.

```python
import numpy as np
import matplotlib.pyplot as plt

# ---- Define some data ----
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

# ---- Plot the data -----

fig, ax = plt.subplots(nrows=2,
                      figsize=(15,15))
CS = ax[0].contour(X, Y, Z)
ax[0].clabel(CS, inline=1, fontsize=10)
ax[0].set_title('Simplest default with labels')
ax[0].axis('equal')

filled = ax[1].contourf(X,Y,Z)
ax[1].set_title('Filled Contour')
fig.colorbar(filled,ax=ax[1])

plt.tight_layout()
```

# Colormaps

Colormaps are just what the name implies. 
You can modify this with the **cmap** argument.
In the comments are some of my favorite colormaps.
As a side note I recommend the linear colormaps. 
If you would like more info on linear colormaps, please refer to the conception of their default colormap *viridis* at matplotlib.org.

```python
import numpy as np
import matplotlib.pyplot as plt

# ---- Create some data ----
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

# ---- Make a filled contour of the data ----
# linear colormaps: viridis,plasma,magma,inferno,cividis
# Diverging: seismic,Spectral
plt.figure(figsize=(15,10))
filled = plt.contourf(X,Y,Z,
                      cmap='magma') # Change colormap with cmap
plt.title('Filled Contour')
plt.colorbar() 
```

# AE and Dst Indices 

```python
import pandas as pd

# ---- Pull in AE and DST indices from a file ----

# pull in data
data_ae=pd.read_csv('data_ae.txt',sep=r'\s+',)
data_dst=pd.read_csv('data_dst.txt',sep=r'\s+',)

# cut data to the day of interest
data_ae=data_ae[data_ae['DATE'].str.match(r'2013-08-(19|2[0-3])')]
data_dst=data_dst[data_dst['DATE'].str.match(r'2013-08-(19|2[0-3])')]
 
#reset the index for plotting
data_ae.reset_index(drop=True,inplace=True)
data_dst.reset_index(drop=True,inplace=True)

#bring the time stamps down to the hour
data_ae['TIME']=data_ae['TIME'].str[0:2]+data_ae['TIME'].str[3:5]
data_dst['TIME']=data_dst['TIME'].str[0:2]+data_dst['TIME'].str[3:5]
```

```python
# ---- Plot Data ----
import matplotlib.pyplot as plt

# set default plot
plt.style.use('default')
fig, axes = plt.subplots(nrows=2,ncols=1,
                        figsize=(15,5))

axes[0].plot(data_ae['AE'])
axes[0].set_title('AE')
axes[0].set_xlim(0,len(data_ae['AE']))

axes[1].plot(data_dst['DST'])
axes[1].set_title('DST')
axes[1].set_xlim(0,len(data_dst['DST']))
 

# ---- Robustly replace labels in the plot ----

# auto extract auto from plot (index in pandas object)
locs=axes[1].get_xticks()

# create empty labels to populate for replacing old labels
hr_labels = []

# run a loop to replace labels with their corresponding time stamp
# exclude the very last item (labels[-1]) because of index-to-data mismatch
for loc in locs[0:-1]:
    temp = data_ae.iloc[int(loc)]['TIME']
    hr_labels.append(temp)
    
# this last label will be added manually as a data point doesn't exist for it
# we append the same as the first because the last point is the beginning of a new day
hr_labels.append(hr_labels[0])
 
# Manually create a date label and an arbitrary label for fun
date_labels=['Aug19','','Aug20','Aug21','Aug22','Aug23','Aug24']
arbitrary=['0','1','2','3','4','5','6']

# Use list comprehension to combine the labels with a newline between each label
combined_labels = ['{}\n{}\n{}'.format(a,b,c) for (a,b,c) in zip(hr_labels,date_labels,arbitrary)]

# Erase the first subplot's labels
axes[0].set_xticklabels([])
# Formality. Since label locations are the same we don't need this line
axes[1].set_xticks(locs)
# Set the bottom subplot's labels
axes[1].set_xticklabels(combined_labels)

# set minor ticks
for ax in axes:
    ax.tick_params(direction='in',length=6)

axes[1].set_xlabel('hour\nDate\nArbitrary Number')
    
plt.show()
```

# MagEIS Data

```python
import numpy as np
import matplotlib.pyplot as plt 

# ---- Load Van Allen Probe MagEIS Electron Data ----
# August 21, 2013 18:00 ~ 20:00
epoch = np.load('rbspb_mageis_20130821_epoch.npy')
pa = np.load('rbspb_mageis_20130821_pitch_angle.npy')
counts = np.load('rbspb_mageis_20130821_counts.npy')
lshell = np.load('rbspb_mageis_20130821_L.npy')
mlt = np.load('rbspb_mageis_20130821_MLT.npy')

# Hard Coded Energy Variables for convenience
energylabels=['31.9 keV',
              '54.5 keV',
              '75.2 keV',
              '101.6 keV',
              '132.4 keV']

nenergies=len(energylabels)
# ---- Make our plot ----
 
# make the same amount of subplots as energies
fig,axes = plt.subplots(nrows=nenergies,
                        figsize=(15,8))

# Plot all energies and counts using a loop
for en in range(nenergies):
    im = axes[en].pcolormesh(  
        epoch,pa,
        counts[:,:,en].transpose(), 
        vmin=0,vmax=80000) 
    #Set Titles 
    axes[en].set_title(energylabels[en])
    axes[en].set_ylabel(r'$\alpha$ [deg]')
    

    
# ---- Modify Ticks and Tick Labels ----

# Erase all tick labels for every subplot except the last
for ax in axes[0:nenergies-1]:
    ax.set_xticklabels([])
    
# Fix reformat the label locations (label locs wrong without adjustment because of epoch)
label_locations = np.linspace(epoch[0],epoch[-1],num=5)
    
# Make our own labels. Hours are hardcoded for convenience
hr_labels = ['18:00','18:30','19:00','19:30','20:00'] 
l_labels=[]
mlt_labels=[]

# populate the labels excluding the last point for out of bounds error
for labeln in np.linspace(0,len(epoch)-1,num=5).astype(int):
    l_labels.append(lshell[labeln])
    mlt_labels.append(mlt[labeln])
    
# combine all labels and format the numbers to only have 1 decimal    
combined_labels = ['{:.1f}\n{:.1f}\n{}'.format(a,b,c) 
                for (a,b,c) in zip(l_labels,mlt_labels,hr_labels)]

# Change all tick locations
for ax in axes:
    ax.set_xticks(label_locations)

# Set the combined labels
axes[-1].set_xticklabels(combined_labels)

# change the x_label coordinate for clarity
axes[-1].xaxis.set_label_coords(-0.1,-0.1)
axes[-1].set_xlabel('L-shell\nMLT\nhr:mm')

# tighten up the layout before adding colorbar, otherwise figure is wrong
fig.tight_layout()    

# scale the right side of the subplot down to make room for the colorbar
fig.subplots_adjust(right=0.8)  
# add the axes for the colorbar at a defined location and size [x,y,w,h]
cbar_ax = fig.add_axes([0.85,0.10,0.01,0.8])
# add the colorbar
cbar = fig.colorbar(im,cax=cbar_ax,aspect=5)
# set colorbar label
cbar.set_label('Counts',fontsize=16)
 
fig.show()  
```
